```python
#distribution plot


plt.figure(figsize=(10,15))
a = 3
b = 2
c = 1
for i in num_cols:
    plt.subplot(a,b,c)
    sns.distplot(sales[i])
    plt.title(i)
    c+=1
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize= (10,15))
a =3 
b =2
c =1
for i in num_cols:
    plt.subplot(a,b,c)
    sns.histplot(sales[i])
    plt.title(i)
    c+=1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,15))
a = 3
b = 2
c = 1
for i in num_cols:
    plt.subplot(a,b,c)
    sns.scatterplot(sales[i])
    plt.title(i)
    c+=1
plt.tight_layout()
plt.show()
```
